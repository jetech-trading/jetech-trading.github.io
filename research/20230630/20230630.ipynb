## MAB(Multi Armed Bandits)

강화학습의 고전적 문제 중 하나로 일정 확률에 의해 보상을 얻을 수 있는 경우 선택을 최적화 하는 문제입니다.



시간의 흐름에 따라 보상을 최대화하는 방법을 찾아내는게 중요하고

컨텐츠들의 추천 알고리즘도 MAB에 기반해서 동작합니다.

유튜브의 알고리즘도 원리 자체는 MAB에 근간을 두고있을거라 생각이듭니다.

공통적으로 강화학습 문제에는 Exploitation(활용)/Exploration(탐험) trade-off 문제가 중요합니다.

활용이란것은 내가 경험에 기반하여 학습한 데이터로 선택에 활용하는것,

탐험은 내가 알고있는것들과 반해 모험적인 선택을 해보는것입니다.



image.png


흔히 우리가 자주 다니던 식당과 새로운 식당에 가보는것과 비유하곤합니다.

강화학습에서 Exploitation/Exploration을 어떻게 다루느냐에 따라 여러 방법들이 존재하는데

오늘은 대표적으로 알려진 4가지 정도를 살펴보고

추천에 사용되는 이런 강화학습 로직을

트레이딩에 어떻게 적용할지 이야기해보도록 하겠습니다.


- e-greedy
- softmax exploration
- UCB(Upper confidence Bound) 알고리즘
- Thompson Sampling

````python
import numpy as np
import gym
from gym import spaces
from gym.utils import seeding


class BanditEnv(gym.Env):
    """
    Bandit environment base to allow agents to interact with the class n-armed bandit
    in different variations

    p_dist:
        A list of probabilities of the likelihood that a particular bandit will pay out
    r_dist:
        A list of either rewards (if number) or means and standard deviations (if list)
        of the payout that bandit has
    """
    def __init__(self, p_dist, r_dist):
        if len(p_dist) != len(r_dist):
            raise ValueError("Probability and Reward distribution must be the same length")

        if min(p_dist) < 0 or max(p_dist) > 1:
            raise ValueError("All probabilities must be between 0 and 1")

        for reward in r_dist:
            if isinstance(reward, list) and reward[1] <= 0:
                raise ValueError("Standard deviation in rewards must all be greater than 0")

        self.p_dist = p_dist
        self.r_dist = r_dist

        self.n_bandits = len(p_dist)
        self.action_space = spaces.Discrete(self.n_bandits)
        self.observation_space = spaces.Discrete(1)

        self._seed()

    def _seed(self, seed=None):
        self.np_random, seed = seeding.np_random(seed)
        return [seed]

    def step(self, action):
        assert self.action_space.contains(action)

        reward = 0
        done = True

        if np.random.uniform() < self.p_dist[action]:
            if not isinstance(self.r_dist[action], list):
                reward = self.r_dist[action]
            else:
                reward = np.random.normal(self.r_dist[action][0], self.r_dist[action][1])

        return 0, reward, done, {}

    def reset(self):
        return 0

    def render(self, mode='human', close=False):
        pass

class BanditTenArmedGaussian(BanditEnv):
    """
    10 armed bandit mentioned on page 30 of Sutton and Barto's
    [Reinforcement Learning: An Introduction](https://www.dropbox.com/s/b3psxv2r0ccmf80/book2015oct.pdf?dl=0)

    Actions always pay out
    Mean of payout is pulled from a normal distribution (0, 1) (called q*(a))
    Actual reward is drawn from a normal distribution (q*(a), 1)
    """
    def __init__(self, bandits=10):
        p_dist = np.full(bandits, 1)
        r_dist = []

        for _ in range(bandits):
            r_dist.append([np.random.normal(0, 1), 1])

        BanditEnv.__init__(self, p_dist=p_dist, r_dist=r_dist)
````

## e-greedy policy

````python
num_rounds = 20000
count = np.zeros(10)
sum_rewards = np.zeros(10)
Q = np.zeros(10)

def e_greedy(eps):
    rand = np.random.random()
    if rand < eps:
        action = env.action_space.sample()
    else:
        action = np.argmax(Q)
    return action
````

e-greedy 알고리즘에서는 랜덤한 확률 중 epsilon확률 안에서는 랜덤으로 행동하고
```python
rand = np.random.random()
|if rand < eps:
    action = env.action_space.sample()
```


그 외에는 가지고있는 행동가치함수 Q에서 최적의 행동을 추출합니다

```python
action = np.argmax(Q)
```

````python
env.reset()
for i in range(num_rounds):
    arm = e_greedy(0.5)
    observation, reward, done, info = env.step(arm)
    count[arm] += 1
    sum_rewards[arm] += reward
    Q[arm] = sum_rewards[arm]/count[arm]
print(f'Optimal arm : {np.argmax(Q)}')
````

## Softmax Exploration

````python
num_rounds = 20000
count = np.zeros(10)
sum_rewards = np.zeros(10)
Q = np.zeros(10)

def softmax(tau):
    total = sum([np.exp(val/tau) for val in Q])
    probs = [np.exp(val / tau) / total for val in Q]
    threshold = np.random.random()
    cumulative_prob = 0.0
    for i in range(len(probs)):
        cumulative_prob += probs[i]
        if cumulative_prob > threshold:
            return i
````

소프트맥스 탐험에서는 최적밴딧을 위해 볼츠만 탐험을 합니다.

e-greedy 정책에선 모든 레버를 동등하게 고려했지만

소프트맥스 탐험에서는 레버를 확률적으로 선택합니다.

tau는 temperature factor로 얼마나 많은 수의 레버를 임의로 탐색할 수 있는지를 나타냅니다.

tau가 크면 모든 레버를 동등하게 탐색하고

작으면 높은 보상을 주는 레버만 선택되게됩니다.

````python
env.reset()
for i in range(num_rounds):
    arm = softmax(0.5)
    observation, reward, done, info = env.step(arm)
    count[arm] += 1
    sum_rewards[arm] += reward
    Q[arm] = sum_rewards[arm]/count[arm]
print('The optimal arm is {}'.format(np.argmax(Q)))
````

## UCB

위 두 가지방법은 확률에 따라 임의의 행동을 탐험할 수 있었습니다.

Exploration은 다양한 행동을 찾아가는덴 유용하지만

어느정도 학습된 이후에는 무작위성이 강해지게 됩니다.

신뢰상한알고리즘은 불확실성에 대해 낙관적 대응을 하는 알고리즘입니다.

UCB 알고리즘은 신뢰구간에 기반하여 행동들을 판단하고

행동들에 대한 보상의 평균을 계산하여 높은 평균 보상값을 주는 행동을 하는것을 택하는 알고리즘입니다.



````python
# UCB
num_rounds = 20000
count = np.zeros(10)

sum_rewards = np.zeros(10)

Q = np.zeros(10)

def UCB(iters):
    ucb = np.zeros(10)
    if iters < 10:
        return
    else:
        for arm in range(10):
            upper_bound = np.sqrt((2*np.log(sum(count))) / count[arm])
            ucb[arm] = Q[arm] + upper_bound
        return np.argmax(ucb)

env.reset()
for i in range(num_rounds):
    arm = UCB(i)
    if arm is None:
        continue
    observation, reward, done, info = env.step(arm)
    count[arm] += 1
    sum_rewards[arm] += reward
    Q[arm] = sum_rewards[arm] / count[arm]
print('The optimal arm is {}'.format(np.argmax(Q)))
````

## Thompson Sampling

TS는 prior distribution에 기반한 확률적 알고리즘입니다.

각 k 행동에 대한 값을 추출하여 이를 사전 평균값으로 사용합니다.

가장 높은 평균 사전 확률값을 갖는 레버를 선택하고 행동한 뒤 보상을 얻습니다.

얻은 보상을 이용하여 사전 분포를 업데이트합니다.

위 과정을 진행하면 사전분포가 참 분포에 점점 가까워집니다.



````python

num_rounds = 20000

count = np.zeros(10)
sum_rewards = np.zeros(10)

Q = np.zeros(10)
# +1 reward count
alpha = np.ones(10)

# -1 reward count
beta = np.ones(10)

def ts(alpha, beta):
    samples = [np.random.beta(alpha[i]+1, beta[i]+1) for i in range(10)]
    return np.argmax(samples)

env.reset()
for i in range(num_rounds):
    arm = ts(alpha, beta)
    observation, reward, done, info = env.step(arm)
    count[arm] += 1
    sum_rewards[arm] += reward
    Q[arm] = sum_rewards[arm] / count[arm]
    if reward > 0:
        alpha[arm] += 1
    else:
        beta[arm] += 1

print('The optimal arm is {}'.format(np.argmax(Q)))
````
